# **Validações do churn predito - Batch**

O percentual de clientes classificados como risco alto faz sentido?

```python
import pandas as pd

df = pd.read_csv("../predictions/churn_predicoes.csv")

df["faixa_risco"].value_counts(normalize=True).round(2)
```

As faixas realmente representam níveis crescentes de risco?

```python
df.groupby("faixa_risco")["probabilidade_churn"].agg(["mean", "min", "max"]).round(2)
```

O modelo está apontando risco onde o EDA mostrou risco?

```python
pd.crosstab(
    df["Contract"],
    df["faixa_risco"],
    normalize="index"
)
```

As validações do churn predito confirmam a coerência
do modelo em produção. Clientes com contratos mensais concentram-se
majoritariamente nas faixas de risco alto e muito alto, enquanto contratos
bianuais apresentam predominância de risco baixo. Esse comportamento
replica os padrões observados na análise exploratória e valida a utilização
das faixas de risco como instrumento de priorização para ações de retenção.
